## 1. Σ, run three ways

$\sum_{i=1}^{5} i^2$ as the literal loop, the one-liner, and numpy. Same Σ, three dialects —
and one trap: Σ's top limit is **inclusive**, so `range` needs `n + 1`.

```python
# dialect 1: the literal translation
total = 0
for i in range(1, 6):          # range(1, 6) gives 1,2,3,4,5 — stops BEFORE 6
    total = total + i**2
print(total)

# dialect 2: the one-liner
print(sum(i**2 for i in range(1, 6)))

# dialect 3: numpy — no visible loop, same Σ underneath
print(np.sum(np.arange(1, 6)**2))
```

## 2. Double Σ = nested loops

$\sum_{i=1}^{2}\sum_{j=1}^{2} w_{ij}$ — outer loop over rows, inner loop over columns.
Read the indentation like the subscripts.

```python
w = [[1, 2],
     [3, 4]]                    # w[i][j]: row i, column j (from 0 in Python)

total = 0
for i in range(2):              # for each row...
    for j in range(2):          # ...for each column in that row
        total = total + w[i][j]
print(total)                     # matches your problem 10?

print(np.sum(w))                 # numpy runs both loops in one call
```

## 3. 🌀 A Σ that draws

So far Σ has only made single numbers. But feed it *waves* and a Σ can build shapes.
Watch this sum of smooth sine waves —

$$\sum_{k=1}^{K} \frac{\sin\big((2k-1)\,x\big)}{2k-1}$$

— construct a **square wave**, something with hard corners, out of curves with none.
More terms in the Σ, sharper the corners:

```python
xs = np.linspace(0, 4 * np.pi, 2000)

plt.figure(figsize=(10, 5))
for K, colour in [(1, "#c7d2fe"), (3, "#818cf8"), (30, "#1a1a2e")]:
    wave = np.zeros_like(xs)
    for k in range(1, K + 1):                     # the Σ, as the loop you now read fluently
        wave = wave + np.sin((2*k - 1) * xs) / (2*k - 1)
    plt.plot(xs, wave, color=colour, linewidth=2, label=f"K = {K} term(s)")

plt.title("A \u03a3 of smooth waves building a square wave")
plt.legend()
plt.show()
```

That's a **Fourier series** — the mathematics inside JPEG images, MP3 audio, and the
curriculum's grand finale (epicycles that can draw *anything*). You just read its Σ like a
for-loop, because now that's what Σ is to you.
