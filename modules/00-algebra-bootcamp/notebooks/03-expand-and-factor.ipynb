## 2. Proof by picture: two shapes, ONE curve

If $x^2 + 6x + 8$ and $(x+2)(x+4)$ are really the same expression, then plotted as
functions they must draw **the same curve** — not close, identical. Let's test that claim
at a thousand points.

```python
xs = np.linspace(-6, 2, 1000)              # 1000 test values of x
expanded = xs**2 + 6*xs + 8
factored = (xs + 2) * (xs + 4)

# np.allclose asks: are ALL 1000 pairs of numbers equal (allowing float fuzz)?
print("identical at all 1000 points?", np.allclose(expanded, factored))
```

## 3. 🌀 Why the factored shape is a superpower

Here's the curve. The **factored form reads its secrets straight off**: $(x+2)(x+4)$
announces "I touch zero at $x=-2$ and $x=-4$" before you compute anything. The expanded
form $x^2+6x+8$ hides them completely. Same curve — but one costume tells you where it
lands.

```python
ax = plot_functions({
    "y = x\u00b2 + 6x + 8  =  (x+2)(x+4)": lambda x: x**2 + 6*x + 8,
}, xlim=(-7, 3), ylim=(-4, 12))

for root in (-2, -4):                       # the roots, read off the factored form
    ax.plot(root, 0, "ko", markersize=9)
ax.annotate("(x+2) says: zero here", (-2, 0), xytext=(-1.2, 4),
            fontsize=11, arrowprops=dict(arrowstyle="->"))
ax.annotate("(x+4) says: zero here", (-4, 0), xytext=(-6.8, 6),
            fontsize=11, arrowprops=dict(arrowstyle="->"))
ax.set_title("Factoring = making the roots visible")
plt.show()
```

*Solving* an equation and *finding where a curve crosses zero* are the same act — and in
Module 3, "where does this curve bottom out?" becomes the question that trains every
neural network. Reshaping expressions until answers become visible is the whole game.
